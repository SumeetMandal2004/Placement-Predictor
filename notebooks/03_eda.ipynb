```python
# Lesson 3: Exploratory Data Analysis (EDA)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/processed/placement_clean.csv')

# Set default style for all charts
sns.set_theme(style='whitegrid')

print(f"Shape: {df.shape}")
```

```python
plt.figure(figsize= (6,4))

sns.countplot(data=df, x='status', hue='status', palette='Set2', legend=False)

plt.title('Placement Status Distribution')
plt.xlabel('Students')
plt.ylabel('Number of Students')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))

sns.countplot(data=df, x='gender', hue='status', palette='Set2')

plt.title('Gender Vs Placement Status')
plt.xlabel('Gender')
plt.ylabel('Number of Students')

plt.tight_layout()
plt.show()
```

```python
gender_placement= df.groupby('gender')['status'].value_counts(normalize=True)*100
print(gender_placement.round(2))
```

```python
plt.figure(figsize=(6,4))

sns.countplot(data=df, x = 'workex', hue='status', palette='Set2')

plt.title('Work Experience Vs Placement Status')
plt.ylabel("Number of Students")
plt.xlabel("Work Experience")

plt.tight_layout()
plt.show()
```

```python
workex_placement = df.groupby('workex')['status'].value_counts(normalize=True)*100
print(workex_placement.round(2))
```

```python
plt.figure(figsize=(6,4))

sns.boxplot(data=df, x='status', y='mba_p', palette='Set2', hue='status', legend=False)

plt.title("MBA Percentage VS Placement Status")
plt.xlabel("Placement Status")
plt.ylabel("MBA Percentage")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))

sns.boxplot(data=df, x='status', y='ssc_p', palette='Set2', hue='status', legend=False)

plt.title("10th Percentage VS Placement Status")
plt.ylabel("10th Percentage")
plt.xlabel("Placement Status")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))

sns.boxplot(data=df, x='status', y='degree_p', palette='Set2', hue='status', legend=False)

plt.title('Degree Percentage vs Placement Status')
plt.xlabel('Placement Status')
plt.ylabel('Degree Percentage')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))

sns.countplot(data=df, x='degree_t', hue='status', palette='Set2')

plt.title('Degree Type vs Placement Status')
plt.xlabel('Degree Type')
plt.ylabel('Number of Students')

plt.tight_layout()
plt.show()
```

```python
df.groupby('degree_t')['status'].value_counts(normalize=True)*100
```

```python
# Correlation Heatmap

plt.figure(figsize=(8,6))

#Selecting only numerical columns
numerical_df = df.select_dtypes(include='number')

#Calculating correaltion
correlation = numerical_df.corr()

# sns.heatmap(correlation, annot=True, fmt='.2f', cmap ='coolwarm', linewidths=0.5)
sns.heatmap(correlation, 
            annot=True,        # Show numbers inside boxes
            fmt='.2f',         # Round to 2 decimal places
            cmap='coolwarm',   # red = positive correlation, blue = negative
            linewidths=0.5)    # Lines between boxes

plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```
